```python
awards_list = []  # 빈 리스트 생성

# url 설정
for year in years_list: 
    
    url = f'https://www.oscars.org{year}'   # 사이트 불러오는 반복문
    
    resp = requests.get(url) 
    
    html = BeautifulSoup(resp.text, 'html.parser')
    
    print('*' * 30, html.select('span.date-display-single')[0].text, '*' * 30)  # 년도 구분문
        
    # 차단막는 코드
    seed = np.random.randint(100)
    np.random.seed(seed)
    a = np.random.randint(5) + round(np.random.random(), 2)
    time.sleep(a)
    
    dic = {}  
    winner_list = []  # 수상자 리스트
    award_list = []  # 상 이름 리스트
    
    
    dic['year'] = html.select('span.date-display-single')[0].text   # 년도 반복문
    print('년도 :', dic['year'])
    
    
    #위치, 날씨 반복문  
    dic['location'] = html.select('div.field-content')[0].text
    dic['date'] = html.select('div.field-content')[1].text.split(',')[1]
    print('위치 :', dic['location'])
    print('날짜 :', dic['date'])
    
    
    for name in html.select('span.field-content')[2::2]:  # 수상자 반복문
        winner_list.append(name.text)
        dic['name'] = winner_list
    print('수상자 :', winner_list)
        
    for award in html.select('div.views-field.views-field-field-governors-award-category'):    # 상 이름 반복문
        award_list.append(award.text)
        dic['award'] = award_list
    print('상이름 :', award_list)
                
                
    awards_list.append(dic)
print()    
print('*' * 30, "END", '*' * 30 )
```
